```python
import networkx as nx
import matplotlib.pyplot as plt

top_nodes = sorted(df['sender_address'].value_counts().index[:100])
top_transactions = df[df['sender_address'].isin(top_nodes) | df['receiver_address'].isin(top_nodes)]

G = nx.DiGraph()
G.add_edges_from(zip(top_transactions['sender_address'], top_transactions['receiver_address']))

pos = nx.spring_layout(G)

node_sizes = [G.degree(node)*10 for node in G.nodes]
edge_widths = [1 for u, v in G.edges]

# Display full addresses
labels = {node: node[:20] for node in G.nodes}

community_colors = []
for node in G.nodes:
    if G.degree(node) > 10:
        community_colors.append('red')
    else:
        community_colors.append('lightblue')

fig, ax = plt.subplots(figsize=(25, 25))

nx.draw_networkx_nodes(G, pos, node_size=node_sizes, node_color=community_colors, ax=ax)
nx.draw_networkx_edges(G, pos, width=edge_widths, edge_color='black', arrowsize=20, ax=ax)
nx.draw_networkx_labels(G, pos, labels=labels, font_size=10, font_color='darkgrey', ax=ax)

ax.axis('off')
plt.show()

```

```python
import plotly.graph_objects as go

node_x = [pos[node][0] for node in G.nodes]
node_y = [pos[node][1] for node in G.nodes]

edge_x = []
edge_y = []
for edge in G.edges:
    source, target = edge
    x0, y0 = pos[source]
    x1, y1 = pos[target]
    edge_x.append(x0)
    edge_x.append(x1)
    edge_x.append(None)
    edge_y.append(y0)
    edge_y.append(y1)
    edge_y.append(None)

node_trace = go.Scatter(
    x=node_x,
    y=node_y,
    mode='markers',
    hoverinfo='text',
    hovertext=[node for node in G.nodes],
    marker=dict(size=node_sizes, color=community_colors)
)

edge_trace = go.Scatter(
    x=edge_x,
    y=edge_y,
    line=dict(width=0.5, color='darkgray'),
    hoverinfo='none',
    mode='lines'
)

fig = go.Figure(data=[node_trace, edge_trace])
fig.update_layout(
    titlefont_size=16,
    showlegend=False,
    hovermode='x',
    margin=dict(b=20, l=5, r=5, t=40),
    xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
    yaxis=dict(showgrid=False, zeroline=False, showticklabels=False)
)

fig.show()

```

```python
from statsmodels.tsa.seasonal import seasonal_decompose, STL
from statsmodels.tsa.stattools import adfuller
```

```python
# Convert timestamp to datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Frequency Analysis
## Daily transaction frequency
daily_freq = df['timestamp'].dt.date.value_counts().sort_index()

## Convert daily_freq to pandas Series with DatetimeIndex
daily_freq_series = pd.Series(daily_freq.values, index=pd.date_range(start=min(df['timestamp']), periods=len(daily_freq), freq='D'))

# Time-Series Analysis
## STL Decomposition (alternative to seasonal_decompose)
stl_decomposition = STL(daily_freq_series).fit()
trend = stl_decomposition.trend
seasonal = stl_decomposition.seasonal
residual = stl_decomposition.resid

plt.figure(figsize=(12, 6))
plt.subplot(411)
plt.plot(daily_freq_series, label='Original')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend, label='Trend')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(seasonal, label='Seasonality')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residual, label='Residuals')
plt.legend(loc='best')
plt.tight_layout()
plt.show()

## Stationarity test using Augmented Dickey-Fuller (ADF) test
result = adfuller(daily_freq_series)
print('ADF Statistic:', result[0])
print('p-value:', result[1])
if result[1] < 0.05:
    print('Reject the null hypothesis - series is likely stationary')
else:
    print('Fail to reject the null hypothesis - series is likely non-stationary')
```
